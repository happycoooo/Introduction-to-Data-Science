```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('./projectData/Levels_Fyi_Salary_Data.csv')
data.head()
```

Initial data processing

```python
data['timestamp'].info()
```

Educational background of data scientists

```python
data['Masters_Degree'] = data.loc[:,"Masters_Degree"].map({ 1:True, 0:False})
data['Bachelors_Degree'] = data.loc[:,"Bachelors_Degree"].map({ 1:True, 0:False})
data['Doctorate_Degree'] = data.loc[:,"Doctorate_Degree"].map({ 1:True, 0:False})
data['Highschool'] = data.loc[:,"Highschool"].map({ 1:True, 0:False})
data['Some_College'] = data.loc[:,"Some_College"].map({ 1:True, 0:False})
data.head()
```

```python
#dict_data = {'time':data.loc[:,'timestamp']}
#time_data = pd.DataFrame(dict_data)
data['timestamp'] = pd.to_datetime(data['timestamp'])
data['Year'] = data['timestamp'].dt.year
data.head()
print(data['Year'].unique())

data_for_2017 = data[data['Year'] == 2017]
data_for_2018 = data[data['Year'] == 2018]
data_for_2019 = data[data['Year'] == 2019]
data_for_2020 = data[data['Year'] == 2020]
data_for_2021 = data[data['Year'] == 2021]
```

```python
Num = job_title['Count'].sum()
data = [2052380,11315444,20435242,7456627,3014264,1972395,185028]
#数据标签
labels = ['none', 'primary', 'junior', 'senior', 'specialties', 'bachelor', 'master']
#各区域颜色
colors = ['red','orange','yellow','green','purple','blue','black']
#数据计算处理
sizes = [data[0]/Num*100,data[1]/Num*100,data[2]/Num*100,data[3]/Num*100,data[4]/Num*100,data[5]/Num*100,data[6]/Num*100]
#设置突出模块偏移值
expodes = (0,0,0.1,0,0,0,0)
#设置绘图属性并绘图
plt.pie(sizes,explode=expodes,labels=labels,shadow=True,colors=colors)
## 用于显示为一个长宽相等的饼图
plt.axis('equal')
plt.show()
```
