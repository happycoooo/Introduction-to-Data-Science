```python
# Remove the star_words
def remove_star_words(s):
    return re.sub('[^\s]*[\*]+[^\s]*', "", s)

data_scientist['Clean_Job_Descriptions'] = data_scientist['Job Description'].map(remove_star_words)
```

```python
# Remove the urls first - Anything that has .com, .co.uk or www. is a url!
def remove_urls(s):
    s = re.sub('[^\s]*.com[^\s]*', "", s)
    s = re.sub('[^\s]*www.[^\s]*', "", s)
    s = re.sub('[^\s]*.co.uk[^\s]*', "", s)
    return s

data_scientist['Clean_Job_Descriptions'] = data_scientist['Clean_Job_Descriptions'].map(remove_urls)
```

```python
# Remove the punctuations
from string import punctuation

def remove_punctuation(s):
    global punctuation
    for p in punctuation:
        s = s.replace(p, '')
    return s

data_scientist['Clean_Job_Descriptions'] = data_scientist['Clean_Job_Descriptions'].map(remove_punctuation)
```

```python
# Convert to lower case
data_scientist['Clean_Job_Descriptions'] = data_scientist['Clean_Job_Descriptions'].map(lambda x: x.lower())
```

```python
# define a function for parts of speech tagging
# make a corpus of all the words in the job description
corpus = " ".join(data_scientist['Clean_Job_Descriptions'].tolist())

# This is the NLTK function that breaks a string down to its tokens
tokens = word_tokenize(corpus)

# Get the parts of speech tag for all words
answer = nltk.pos_tag(tokens)
answer_pos = [a[1] for a in answer]

# print a value count for the parts of speech
all_pos = pd.Series(answer_pos)
all_pos.value_counts().head()
```

How does the frequency change if we exclude stopwords?
Stopwords are words such as I, me, myself, they, and, etc. that are useful for meaningful sentence structure but do not really add value to the prediction task at hand. This is because they appear with equal probability in both high salary and low salary job descriptions. To make computations faster, we remove these stopwords before we go into the prediction algorithm!

Approach：

1.Make a list of stopwords from the nltk library.
2.Remove stopwords from each of the descriptions.
3.Apply the pos_tagger again.

```python
from nltk.corpus import stopwords
en_stopwords = stopwords.words('english')

# define a function to remove stopwords from descriptions
def remove_stopwords(s):
    global en_stopwords
    s = word_tokenize(s)
    s = " ".join([w for w in s if w not in en_stopwords])
    return s

# Create a new column of descriptions with no stopwords
data_scientist['Clean_Job_Descriptions_no_stop'] = data_scientist['Clean_Job_Descriptions'].map(remove_stopwords)

# make a corpus of all the words in the job description
corpus = " ".join(data_scientist['Clean_Job_Descriptions_no_stop'].tolist())

# This is the NLTK function that breaks a string down to its tokens
tokens = word_tokenize(corpus)

answer = nltk.pos_tag(tokens)
answer_pos = [a[1] for a in answer]

all_pos = pd.Series(answer_pos)
all_pos.value_counts().head()
```

What are the 10 most common words in the job description data?

```python
# import the necessary functions from the nltk library
from nltk.stem.wordnet import WordNetLemmatizer
lmtzr = WordNetLemmatizer()

# prepare corpus from the descriptions that dont have stopwords
corpus = " ".join(data_scientist['Clean_Job_Descriptions_no_stop'].tolist())

#tokenize words
tokenized_corpus = nltk.word_tokenize(corpus)

# lemmatize these tokens
lemmatized_tokens = [lmtzr.lemmatize(token) for token in tokenized_corpus]

# word frequencies for the lemmatized tokens
fd = nltk.FreqDist(lemmatized_tokens)

# get the top words
top_words = []
for key, value in fd.items():
    top_words.append((key, value))

# sort the list by the top frequencies
top_words = sorted(top_words, key = lambda x:x[1], reverse = True)

# keep top 10 words only
top_words = top_words[:10]

top_words
```

```python
#Split up the salary estimate into minimum and maximum salary:
test = data_scientist['Salary Estimate'].str.split('-', expand=True)
data_scientist['MinSalary'] = pd.to_numeric(test[0].str.extract('(\d+)', expand=False))
data_scientist['MaxSalary'] = pd.to_numeric(test[1].str.extract('(\d+)', expand=False))

data_scientist.head()
```

```python
def plot_salary(y, salary, amt=30, title=''):
    l = list(data_scientist[y].value_counts().iloc[:amt].index)
    temp = data_scientist[data_scientist[y].isin(l)]
    
    salary_order = temp.groupby(y)[salary].mean().sort_values(ascending=False).index.values
    
    chart = sns.catplot(data=temp, y=y, x=salary, kind='box', 
                    order=salary_order, 
                    height=10, aspect=1)
    
    plt.title(title)
#     chart.set_xticklabels(rotation=45)

def least_freq(group, target, min_counts=3, max_counts=11):
    counts = Counter(data_scientist[group])
    sal = data_scientist[data_scientist[group]\
                   .isin([x for x in counts if (counts[x] > min_counts & counts[x] < max_counts)])]\
    .groupby(group)[target].mean().reset_index()

    return sal.sort_values(by=(target),ascending=False).head(10)
```

```python
plot_salary('Job Title', 'MinSalary', amt=12, title='Minimum Salary by Job Title')
```

```python
plot_salary('Job Title', 'MaxSalary', amt=12, title='Maximum Salary by Job Title')
```

```python
def by_multiple(group, target, min_counts=11):
    counts = Counter(data_scientist[group[0]])
    sal = data_scientist[data_scientist[group[0]]\
               .isin([x for x in counts if counts[x] > min_counts])]\
               .groupby(group)[target[0]].mean().reset_index()

    return sal.sort_values(by=(target),ascending=False).head(10)
```
