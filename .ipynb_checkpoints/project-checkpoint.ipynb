 #                                                1-Data Preprocessing  

```python
#导入参与该数据科学领域公司培训的申请者信息，处理缺失数据
data =pd.read_csv('./projectData/aug_train.csv')
data['company_size'].fillna('0',inplace=True)
data['company_type'].fillna('Unknown',inplace=True)
data['major_discipline'].fillna('Unknown',inplace=True)
data['gender'].fillna('Not provided',inplace=True)
data.dropna(inplace=True)
data['relevent_experience'] = data.loc[:,"relevent_experience"].map({ 'Has relevent experience':True, 'No relevent experience':False})
data['company_size'] = data['company_size'].apply(lambda x: '10-49' if x == '10/49' else x)
data['experience'] = data['experience'].apply(lambda x: '0' if x == '<1' else x)
data['experience'] = data['experience'].apply(lambda x: '20' if x == '>20' else x)
data['experience'] = data['experience'].astype(str).astype(int)
data.head()
```
